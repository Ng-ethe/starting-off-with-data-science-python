```python
time = np.arange(0., 5., 0.2)
time
```

```python
time2 = time**2
```

```python
time3 = np.exp(time)
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
```

```python
plt.plot(time, 'b', label = 'Regular')
plt.plot(time2, 'r',label = 'Squared')
plt.plot(time3, 'k', label = 'Expotential')
plt.title('Time Taken for python processes', fontdict={'fontsize': 20})
plt.legend(loc = 'best') # This will prove the Graph legend
plt.show()
```

## We can also add the xaxis and yaxis labels using `plt.xlabel` and `plt.ylabel`

```python
plt.plot(time, 'b', label = 'Regular')
plt.plot(time2, 'r',label = 'Squared')
plt.plot(time3, 'k', label = 'Expotential')
plt.title('Time Taken for python processes', fontdict={'fontsize': 20})
plt.xlabel("Number of iterations", fontdict= {'fontsize': 15})
plt.ylabel("Variations ", fontdict= {'fontsize': 15})
plt.legend(loc = 'best') # This will prove the Graph legend with the best location for it
plt.show()
```

```python
# to Clean the axis we can use xticks or yticks to clear the axis
plt.plot(time, 'b', label = 'Regular')
plt.plot(time2, 'r',label = 'Squared')
plt.plot(time3, 'k', label = 'Expotential')
plt.title('Time Taken for python processes', fontdict={'fontsize': 20})
plt.xlabel("Number of iterations", fontdict= {'fontsize': 15})
plt.ylabel("Variations ", fontdict= {'fontsize': 15})
plt.legend(loc = 'best') # This will prove the Graph legend with the best location for it
plt.xticks([])
plt.show()
```

## we can change the `plt.xticks()` values on the x axis also

```python

plt.plot(time, 'b', label = 'Regular')
plt.plot(time2, 'r',label = 'Squared')
plt.plot(time3, 'k', label = 'Expotential')

plt.title('Time Taken for python processes', fontdict={'fontsize': 20})

plt.xlabel("Number of iterations", fontdict= {'fontsize': 15})
plt.ylabel("Variations ", fontdict= {'fontsize': 15})

plt.legend(loc = 'best') 

plt.xticks(np.arange(10,25,5))

# plt.yticks(np.arange(10,25,5))

plt.show()
```
